```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

def demo_bayes():
    #1.获取数据
    """subset='all'表示选择要加载的数据集：训练集为“ train”，
    测试集为“ test”，两者都为“ all”，并按随机顺序排序。
    """
    news = fetch_20newsgroups(subset='all')
    #2.划分数据集
    x_train, x_test, y_train, y_test = train_test_split(news.data, news.target, test_size=0.3)
    #3.特征工程
    tf = TfidfVectorizer()#本次特征抽取没有用到停用词

    x_train = tf.fit_transform(x_train)
    # 这里打印出来的列表是：训练集当中的所有不同词的组成的一个列表
    print(tf.get_feature_names())
    #print(x_train.toarray())

    # 不能调用fit_transform（因为需要用到训练集里面的均值与方差，所以直接使用transform就可以直接使用训练集里面的均值与方差）
    x_test = tf.transform(x_test)

    #4.bayes预估器流程
    mlb = MultinomialNB(alpha=1.0)#默认拉普拉斯平滑系数为1

    mlb.fit(x_train, y_train)
    #5.模型评估
    y_predict = mlb.predict(x_test)
    print("预测每篇文章的类别：", y_predict)
    print("真实类别为：", y_test)
    print("预测准确率为：", mlb.score(x_test, y_test))
    return None

demo_bayes()


```
