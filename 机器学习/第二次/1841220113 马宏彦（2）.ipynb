```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

def demo_bayes():
    #1.่ทๅๆฐๆฎ
    """subset='all'่กจ็คบ้ๆฉ่ฆๅ่ฝฝ็ๆฐๆฎ้๏ผ่ฎญ็ป้ไธบโ trainโ๏ผ
    ๆต่ฏ้ไธบโ testโ๏ผไธค่้ฝไธบโ allโ๏ผๅนถๆ้ๆบ้กบๅบๆๅบใ
    """
    news = fetch_20newsgroups(subset='all')
    #2.ๅๅๆฐๆฎ้
    x_train, x_test, y_train, y_test = train_test_split(news.data, news.target, test_size=0.3)
    #3.็นๅพๅทฅ็จ
    tf = TfidfVectorizer()#ๆฌๆฌก็นๅพๆฝๅๆฒกๆ็จๅฐๅ็จ่ฏ

    x_train = tf.fit_transform(x_train)
    # ่ฟ้ๆๅฐๅบๆฅ็ๅ่กจๆฏ๏ผ่ฎญ็ป้ๅฝไธญ็ๆๆไธๅ่ฏ็็ปๆ็ไธไธชๅ่กจ
    print(tf.get_feature_names())
    #print(x_train.toarray())

    # ไธ่ฝ่ฐ็จfit_transform๏ผๅไธบ้่ฆ็จๅฐ่ฎญ็ป้้้ข็ๅๅผไธๆนๅทฎ๏ผๆไปฅ็ดๆฅไฝฟ็จtransformๅฐฑๅฏไปฅ็ดๆฅไฝฟ็จ่ฎญ็ป้้้ข็ๅๅผไธๆนๅทฎ๏ผ
    x_test = tf.transform(x_test)

    #4.bayes้ขไผฐๅจๆต็จ
    mlb = MultinomialNB(alpha=1.0)#้ป่ฎคๆๆฎๆๆฏๅนณๆป็ณปๆฐไธบ1

    mlb.fit(x_train, y_train)
    #5.ๆจกๅ่ฏไผฐ
    y_predict = mlb.predict(x_test)
    print("้ขๆตๆฏ็ฏๆ็ซ็็ฑปๅซ๏ผ", y_predict)
    print("็ๅฎ็ฑปๅซไธบ๏ผ", y_test)
    print("้ขๆตๅ็กฎ็ไธบ๏ผ", mlb.score(x_test, y_test))
    return None

demo_bayes()


```
